### from obsevation which needs to be confirmed later, the fuel_unit has some Null values(NaN)<br> this will be confirmed below

```python
df_missing_data = df_fuel_data.isnull()
df_missing_data.head(10)
```

## STAGE ONE: DATA WRANGLING, PREPROCESSOR, CLEANING

```python
for column in df_missing_data.columns.values.tolist():
    print(column)
    print(df_missing_data[column].value_counts())
    print("")
```

### from the preprocessing analysis, it can be concluded that the darta contain null data for 180 entries.

## Dealing with the missing data, there are two ways of dealing with it, these are:
### 1. droping the entire  rows of the null value which will have impact on the analysis later on
### 2. filling the missing value with predicted values

### 1. dropping the entire rows containing the NaN

```python
df_drop_missing_values= df_fuel_data.dropna(subset = ["fuel_unit"], axis = 0)
df_drop_missing_values.reset_index(drop =True, inplace = True)
df_drop_missing_values.head()
```

```python
df_drop_missing_values.head()
```

```python
df_fuel_data.isnull().sum()
```

```python
df_drop_missing_values.info()
```

### 2. Replacing the missing values with mode of the data

```python
df_mode = df_fuel_data["fuel_unit"].value_counts().idxmax()
df_mode
```

```python
import numpy as np
df_fill_missing = df_fuel_data.replace(np.NaN, df_mode)
df_fill_missing.info()
```
